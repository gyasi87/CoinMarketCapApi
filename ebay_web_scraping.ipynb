```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
```

```python
url = 'https://www.ebay.com/sch/i.html?_from=R40&_nkw=phones&_sacat=0&LH_TitleDesc=0&_pgn=1'
soup=BeautifulSoup(requests.get(url).text,"html.parser")
```

```python
links = []
next_page_url = 'https://www.ebay.com/sch/i.html?_from=R40&_nkw=phones&_sacat=0&LH_TitleDesc=0&_ipg=200'
while True:
    response=requests.get(next_page_url)
    soup=BeautifulSoup(response.text,"lxml") 
    items = soup.findAll('a', class_='s-item__link')
    for i in items:
        links.append(i['href'])
        print(i['href'])
    next_page = soup.find('a',class_='pagination__next icon-link')
    if next_page['type'] =='next':
        next_page_url=next_page['href']
        print(next_page_url)
    else:
        break
```

```python
title = []
condition = []
price = []
num_available = []
num_sold = []
returns = []
watchers = []
seller = []
num_of_reviews = []
perc_of_reviews = []
num_product_rating = []
product_rating = []
sold_last_24h = []
```

```python
for page in links:
    response=requests.get(page)
    soup=BeautifulSoup(response.content,'html.parser')

    try:
        title.append(soup.find('h1').next.next.next)
    except:
        title.append('None')
    try:
        condition.append(soup.find('div',class_='u-flL condText').text)
    except:
        condition.append('None')
    try:
        price.append(soup.find('span',class_='notranslate')['content'])
        
    except:
        price.append('None')
    try:
        num_available.append(soup.find('span',id='qtySubTxt').text.strip())
        
    except:
        num_available.append('None')
    try:
        num_sold.append(soup.find_all('span',class_='w2b-head')[0].text)
        
    except:
        num_sold.append('None')
    try:
        returns.append(soup.find_all('span',class_='w2b-head')[1].text)
        print(soup.find_all('span',class_='w2b-head')[1].text)
        
    except:
        returns.append('None')        
    try:
        watchers.append(soup.find_all('span',class_='w2b-head')[2].text)
        print(soup.find_all('span',class_='w2b-head')[2].text)
        
    except:
        watchers.append('None')   
    try:
        seller.append(soup.find('span',class_='mbg-nw').text)
        print(soup.find('span',class_='mbg-nw').text)
        
    except:
        seller.append('None')   
    try:
        num_of_reviews.append(soup.find('span',class_='mbg-l').text.strip().replace('\n',''))
        print(soup.find('span',class_='mbg-l').text.strip().replace('\n',''))
        
    except:
        num_of_reviews.append('None')
    try:
        perc_of_reviews.append(soup.find('div',id='si-fb').text)
        print(soup.find('div',id='si-fb').text)
        
    except:
        perc_of_reviews.append('None') 
    try:
        num_product_rating.append(soup.find('span',id='_rvwlnk').text)
        print(soup.find('span',id='_rvwlnk').text)
        
    except:
        num_product_rating.append('None')      
    try:
        product_rating.append(soup.find('span',class_='reviews-star-rating')['title'])
        print(soup.find('span',class_='reviews-star-rating')['title'])
        
    except:
        product_rating.append('None')              
        print('None')

    try:
        sold_last_24h.append(soup.find('a',class_='vi-txt-underline').text)              
        print(soup.find('a',class_='vi-txt-underline').text)   
    except:
        sold_last_24h.append('None')
        print('None')
       
```

```python
phone_info = {'title':title,
               'condition':condition,
               'price':price,
               'num_available':num_available,
               'num_sold':num_sold,
               'returns':returns,
               'watchers':watchers,
               'seller':seller,
               'num_of_reviews':num_of_reviews,
               'perc_of_reviews':perc_of_reviews,
               'num_product_rating':num_product_rating,
               'sold_last_24h':sold_last_24h} 
```
